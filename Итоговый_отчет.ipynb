# Итоговый отчет

## 1. Hse lyceum makes me cry
---

#### 2. **Описание идеи проекта**
Проект представляет собой классическую консольную игру "Змейка", адаптированную с дополнительными функциями управления. Основная цель игры — управление змейкой, которая перемещается по игровому полю, собирая яблоки. С каждым съеденным яблоком змейка увеличивается в длину, что усложняет управление. Уникальной особенностью данной реализации является использование системы ввода последовательности букв для изменения направления движения змейки. Это добавляет элемент сложности и требует от игрока внимательности и быстрой реакции.

---

#### 3. **Функциональные возможности**
Проект предоставляет следующие возможности для пользователя:

- **Управление змейкой:**
  - Для изменения направления движения змейки игрок должен ввести некую последовательность букв, отображаемую под игровым полем, а затем указать направление (вправо — `r`, влево — `l`, вверх — `u`, вниз — `d`).
  - Змейка автоматически движется в текущем направлении до тех пор, пока игрок не изменит его.

- **Сбор яблок:**
  - На игровом поле случайным образом появляются яблоки. Когда змейка достигает яблока, она увеличивается в длину, а на поле генерируется новое яблоко.

- **Игровой процесс:**
  - Игра продолжается до тех пор, пока змейка не столкнется с границами игрового поля или сама с собой.
  - Чем длиннее, тем лучше
  - После завершения игры пользователю предлагается сыграть снова или выйти.

- **Настройка сложности:**
  - Сложность игры регулируется параметрами `difficulty` (длина последовательности букв для управления) и `ease` (скорость движения змейки).

- **Визуализация:**
  - Игровое поле отображается в консоли, где:
    - `#` — тело змейки.
    - `O` — яблоко.
    - `.` — пустое пространство.
    - Число — голова змейки (отображается текущая длина змейки).

- **Рестарт игры:**
  - После завершения игры пользователь может выбрать, хочет ли он сыграть снова (`y`) или выйти (`n`).

---

#### 4. **Технический стек проекта**
Проект реализован на языке C++ с использованием стандартных библиотек и функций. Основные используемые компоненты:

- **Стандартные библиотеки C++:**
  - `<iostream>` - база
  - `<chrono>` — для управления временем (задержки между движениями змейки).
  - `<thread>` — для создания многопоточности (отдельный поток для обработки ввода пользователя).
  - `<queue>` — для хранения очереди движений сегментов змейки.
  - `<vector>` — для хранения игрового поля и других структур данных.
  - `<sstream>` — для работы со строками.
  - `<limits>` — для очистки буфера ввода.
